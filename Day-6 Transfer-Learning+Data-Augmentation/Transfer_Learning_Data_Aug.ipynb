The Keras deep learning library provides the ability to use data augmentation automatically when training a model.

This is achieved by using the [ImageDataGenerator class.](https://keras.io/api/preprocessing/image/)

First, the class may be instantiated and the configuration for the types of data augmentation are specified by arguments to the class constructor.

A range of techniques are supported, as well as pixel scaling methods. We will focus on five main types of data augmentation techniques for image data; specifically:

 - Image shifts via the width_shift_range and height_shift_range arguments.
 - Image flips via the horizontal_flip and vertical_flip arguments.
 - Image rotations via the rotation_range argument
 - Image brightness via the brightness_range argument.
 - Image zoom via the zoom_range argument.

For example, an instance of the ImageDataGenerator class can be constructed.

  
**Image Augmentation**


Data augmentation is a technique to artificially create new training data from existing training data. This is done by applying domain-specific techniques to examples from the training data that create new and different training examples.


Image data augmentation is perhaps the most well-known type of data augmentation and involves creating transformed versions of images in the training dataset that belong to the same class as the original image.


```python
#Hortizontal or Vertical Augmentation 

from numpy import expand_dims
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import ImageDataGenerator
from matplotlib import pyplot

#Load the image
img = load_img("Elephant.jpeg")

#convert to numpy array
img = img_to_array(img)

#expand the dimention to depict that we have just one sample 
img = np.expand_dims(img,axis = 0)

#Instance of the data augmentation generator
imggen = ImageDataGenerator(width_shift_range= [-200,200])

#Iterator to get the data 
it = imggen.flow(img,batch_size =1)

#generate samoles and plotting the images
for i in range(9):
  pyplot.subplot(330 + 1 + i)
  
  #Generate the batches of images
  batch = it.next()

  #Convert to unsigned integers to plot and see 
  image = batch[0].astype('uint8')

  pyplot.imshow(image)

#Showing the image
pyplot.show()
```

**Transfer Learning + Data Augmentation**

**Model Defenition**

```python
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense,GlobalAveragePooling1D
```

```python
num_classes = 2

model = Sequential()
model.add(ResNet50(include_top = False, pooling = 'avg',weights = 'imagenet'))
model.add(Dense(num_classes,activation = 'softmax'))

#Indicate whether the layer of Resnet should be trained/changed or not 
model.layers[0].trainable = False
```

**Model Compilation**

```python
model.compile(optimizer = 'sgd',loss= 'categorical_crossentropy',metrics = ['accuracy'])
```

```python
from tensorflow.keras.preprocessing.image import load_img
img = load_img("gdrive/MyDrive/images/train/sideways/dog.11103.jpg")
img
```

```python
from tensorflow.keras.applications.resnet50 import preprocess_input
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img
```

```python
image_size = 224
data_generator = ImageDataGenerator(preprocess_input)
```

```python
train_generator = data_generator.flow_from_directory(directory = 'gdrive/MyDrive/images/train',
                                                     target_size = (image_size,image_size),
                                                     batch_size =10,
                                                     class_mode = 'categorical')

validation_generator = data_generator.flow_from_directory(directory = 'gdrive/MyDrive/images/val',
                                                     target_size = (image_size,image_size),
                                                     batch_size =10,
                                                     class_mode = 'categorical')
```

```python
fit_history = model.fit_generator(
    train_generator,
    epochs = 2,
    steps_per_epoch = 22, #-> 220/22 = 10 samples per step, Batch_size =10
    validation_data = validation_generator,
    validation_steps = 1
)
```

```python
filenames = validation_generator.filenames
no_samples = len(filenames)

print(no_samples)
```

```python
#predict my new model results
predict = model.predict_generator(validation_generator,steps = no_samples)
```

```python
pred = np.argmax(predict,axis = 1) #returns the indices of the max values from the array
from sklearn.metrics import confusion_matrix,accuracy_score 
print(confusion_matrix(pred,validation_generator.labels))

accuracy_score(pred,validation_generator.labels)
```

**Data Augmentation**

```python
#Image generator with augmentation
image_size = 224
data_generator_with_aug = ImageDataGenerator(preprocessing_function = preprocess_input,
                                             horizontal_flip = True,
                                             width_shift_range = 0.2,
                                             height_shift_range = 0.2)

train_generator = data_generator_with_aug.flow_from_directory(directory = 'gdrive/MyDrive/images/train',
                                                     target_size = (image_size,image_size),
                                                     batch_size =10,
                                                     class_mode = 'categorical')
```

```python
#Image generator without augmentation
image_size = 224
data_generator_no_aug = ImageDataGenerator(preprocessing_function = preprocess_input)

validation_generator_no_aug = data_generator_no_aug.flow_from_directory(directory = 'gdrive/MyDrive/images/val',
                                                     target_size = (image_size,image_size),
                                                     batch_size =10,
                                                     class_mode = 'categorical')
```

```python
fit_history = model.fit_generator(
    train_generator,
    epochs = 2,
    steps_per_epoch = 22, #-> 220/22 = 10 samples per step, Batch_size =10
    validation_data = validation_generator_no_aug,
    validation_steps = 1
)
```

```python
test_datagen = ImageDataGenerator(preprocess_input)

test_generator = test_datagen.flow_from_directory(directory = 'gdrive/MyDrive/images/val',
                                                     target_size = (image_size,image_size),
                                                    color_mode = 'rgb',
                                                  shuffle = False, 
                                                     batch_size =1,
                                                     class_mode = 'categorical')
```

```python
filenames = test_generator.filenames
no_samples = len(filenames)

print(no_samples)

#predict my new model results
predict = model.predict_generator(test_generator,steps = no_samples)

pred = np.argmax(predict,axis = 1) #returns the indices of the max values from the array
from sklearn.metrics import confusion_matrix,accuracy_score 
print(confusion_matrix(pred,test_generator.labels))

accuracy_score(pred,test_generator.labels)
```
